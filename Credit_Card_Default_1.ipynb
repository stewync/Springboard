### Grouping age into various range

```python
#
def get_age_range(age):
    j = 20
    for i in range(6):
        if age < j+10:
            break
        j = j + 10
    return j
```

```python
file['age_range'] = file['AGE'].apply(lambda x:get_age_range(x))
```

## Spliting data into Train and Test set

```python
X = file.iloc[:,:-2]
Y = file.iloc[:,-2]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2)
```

## Trying various classification algorithms

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(X_train,Y_train)
score = RFC.score(X_test, Y_test)
print("Score : % {} using {}".format(score*100, 'RandomForestClassifier'))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LRC = LogisticRegression(penalty='l1')
LRC.fit(X_train,Y_train)
score=LRC.score(X_test, Y_test)
print("Score : % {} using {}".format(score*100, 'Logistic Regression'))
```

### Gradient Boost Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

GBC = GradientBoostingClassifier(max_depth=4)
GBC.fit(X_train,Y_train)
score = GBC.score(X_test,Y_test)
print("Score : %{} using {}".format(score*100, 'GradientBoostingClassifier'))
```
