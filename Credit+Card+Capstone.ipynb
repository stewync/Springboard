#### Education has values which range from 0 to 6 where (0,5,6 are undocumented )

```python
file['EDUCATION'].value_counts()
```

#### Replace all undocumented values(0,5,6) to 4 -> 'others' 

1 -> Graduate School

2 -> University

3 -> High School

4 -> Others

```python
file['EDUCATION'] = file['EDUCATION'].replace([0,5,6],4)
education_count = file['EDUCATION'].value_counts()

%matplotlib inline
education_count.plot(kind ='bar', title = 'Education')#file.plot(file['EDUCATION'], kind ='bar', legend = None)

```

#### Marriage | Replace undocumented value 0 to 3 (others)
1 -> Married

2 -> Single

3 -> Others

```python
file['MARRIAGE'] = file['MARRIAGE'].replace(0,3)
marriage_count = file['MARRIAGE'].value_counts()
marriage_count.plot(kind = 'bar', title ='MARRIAGE')
```

```python
print(file[['PAY_1','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']].describe())

print(file['PAY_1'].value_counts())
```

#### History of Past Payment file['PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
####  Replace undocumented [-2,0] -> -1 

-1 : Pay duly ||
1 : Payment delay for 1 month ..... 8: Payment delay for 8 month

PAY_1 : Repayment Status in September

PAY_2 : Repayment Status in August

PAY_3 : Repayment Status in July

PAY_4 : Repayment Status in June

PAY_5 : Repayment Status in May

PAY_6 : Repayment Status in April 


```python
for val in file[['PAY_1','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']]:
    file[val] = file[val].replace([-2,0],-1)
```
